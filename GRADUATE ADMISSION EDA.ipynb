# GRADUATE ADMISSION DATASET CLEANING


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv(r"C:\Users\ADMIN\Downloads\Admission_Predict.csv" )
```

```python
dataset.head(20)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
#Removing irrelevant column(s)

dataset.drop(columns=['Serial No.'],inplace=True)
```

```python
#Removing Nulls

dataset.dropna(axis='columns',how='any')
```

# PERFORMING EXPLORATORY DATA ANALYSIS

```python
import seaborn as sns
```

```python
# simple sns visualization

#dataset[['budget', 'gross']] = df[['budget', 'gross']].astype(float)
sns.regplot(x='TOEFL Score',y='CGPA',data=dataset)
#change dot color to enhance visual
sns.regplot(x='TOEFL Score',y='CGPA',data=dataset,scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
 #Looking at correlation
dataset.corr(method='spearman')
#Confirms high correlation between GRE Score and Chance of Admit
```

```python
#Establishing the correlation Matrix
correlation_matrix = dataset.corr(method='spearman')

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for University Admission Features')
plt.xlabel('Admission Features')
plt.ylabel('Admission Features')
plt.show()
```
